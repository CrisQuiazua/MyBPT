## Razones

```python
ejey_OHb=np.log((FusionData['oiii_5007_flux'])/(FusionData['h_beta_flux']))

ejex_NH=np.log((FusionData['nii_6548_flux']+FusionData['nii_6584_flux'])/FusionData['h_alpha_flux'])
ejex_SH=np.log((FusionData['sii_6717_flux']+FusionData['sii_6731_flux'])/FusionData['h_alpha_flux'])
ejex_OH=np.log((FusionData['oi_6300_flux'])/(FusionData['h_alpha_flux']))
```

## Gráficas 2D y 3D - NII:

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(ejex_NH, ejey_OHb, marker=".", color="purple")
# Cuadrado:
ax.set_xlim([-4.5, 3])
ax.set_ylim([-4, 4])
plt.title('BPT Clásico, Gráfico a')
plt.ylabel('log [O III]/ Hβ')
plt.xlabel('log [NII] /Hα')

# Kewley
x1 = np.linspace(-4.5,0.4)  
y1 = 1.19 + (0.61/(x1-0.47))
plt.plot(x1, y1, '-', color='red', lw=2, label= 'Kewley') 


# Kauffmann
x2 = np.linspace(-4.5,0)
y2 = 1.3 + (0.61/(x2-0.05))
plt.plot(x2, y2, '--', color='blue', lw=2, label='Kouffman')

# Línea Seyfert-LINER
x3 = np.linspace(-0.2, 5)
y3 = 1.01 * x3 + 0.48
ax.plot(x3, y3, '-.', color='green', lw=2, label='Seyfert-LINER')

plt.legend()
```

```python
# Crear una figura y los ejes 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar los datos en 3D
ax.scatter(ejex_NH, ejey_OHb, Log_12_25, c='purple', marker='.')

# Etiquetas de los ejes
ax.set_xlabel('ejex_NH')
ax.set_ylabel('ejey_OHb')
ax.set_zlabel('Log_12_25')

# Agregar líneas en los ejes X e Y
ax.plot(ejey_OHb, np.zeros_like(ejey_OHb), zs=0, zdir='y', color='black', linestyle='--', linewidth=2)
ax.plot(np.zeros_like(ejex_NH), ejex_NH, zs=0, zdir='x', color='black', linestyle='--', linewidth=2)

# Ajustar límites de los ejes para centrar la visualización
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 2)
```

```python
# Contar datos positivos en ejey_OHb y ejex_NH
positivos_ejex_NH = np.sum(ejex_NH > 0)
positivos_ejey_OHb = np.sum(ejey_OHb > 0)

print(f"Datos positivos en ejex_NH: {positivos_ejex_NH}")
print(f"Datos positivos en ejey_OHb: {positivos_ejey_OHb}")
```

## Gráficas 2D y 3D - SII:

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(ejex_SH,ejey_OHb, marker="+", color="blue")

# Cuadrado:
ax1.set_xlim([-4, 1])
ax1.set_ylim([-3, 3])
plt.title('BPT Clásico, Gráfico b')
plt.ylabel('log [O III]/ Hβ')
plt.xlabel('log [SII] /Hα')

# Kewley
x1 = np.linspace(-7, 0.3) 
y1 = 1.30 + (0.72/(x1-0.32))
plt.plot(x1, y1, '-' , color='red', lw=2, label= 'Kewley') 


# Seyfert-LINER
x2 = np.linspace(-0.32,4.0)
y2 = 0.76 + (1.89*x2)
plt.plot(x2, y2, '-.', color='green', lw=2, label='Seyfert-LINER')

plt.legend()
```
