```python
# Define una función que tomará una fila y devolverá la etiqueta adecuada
def etiqueta_fila(row):
    etiquetas = [row["PreEtiquetaNII"], row["PreEtiquetaSII"], row["PreEtiquetaOI"]]
    if etiquetas.count(etiquetas[0]) >= 2:
        return etiquetas[0]
    else:
        return 0
```

```python
# Aplica la función a lo largo de las filas del DataFrame
dfTotal["ETIQUETA"] = dfTotal.apply(etiqueta_fila, axis=1)

dfTotal[7405:7411]
```
