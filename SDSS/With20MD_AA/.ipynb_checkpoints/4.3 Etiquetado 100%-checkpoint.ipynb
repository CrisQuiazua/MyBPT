Se hara la condicion:  
    si el valor de la columna “ETIQUETA” es igual a LINER entonces poner de una vez LINER.  
Si no: Si los valores comparados entre las 2 columnas son igual poner el nombre del valor.  
Para todo lo demás poner NaN  


```python
# Definir las condiciones para la nueva columna "EtiqRef"
condiciones = [
    (dfTotal['ETIQUETA'] == 'LINER'),  # Si "ETIQUETA" es igual a "LINER"
    (dfTotal['ETIQUETA'] == dfTotal['Coin100']),  # Si "ETIQUETA" es igual a "Coin100"
]
```

```python
# Definir los valores correspondientes para cada condición
valores = ['LINER', dfTotal['ETIQUETA']]
```

```python
# Aplicar las condiciones y asignar los valores correspondientes
dfTotal['EtiqRef'] = np.select(condiciones, valores, default=np.nan)
```
