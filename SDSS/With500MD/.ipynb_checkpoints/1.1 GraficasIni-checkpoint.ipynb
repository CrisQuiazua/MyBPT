Objetivo:
1. Realizar descarga de datos desde astroquery: objid, ra, dec y los flujos de emisión del BPT clásico, con los 20.000 datos de galaxias del SDSS.
2. Realizar BPT clásico: Razones de líneas y gráficas primeras.
3. Dibujar líneas de separación de fuente de ionización según ecuaciones de: Kewley, Kauffmann y Cid-Fernandes (Mismas ecuaciones).

```python
import numpy as np
import pandas as pd
import time
from datetime import date
import warnings
warnings.filterwarnings 
import astropy.units as u
from astropy.table import Table
from astroquery.sdss import SDSS
import matplotlib.pyplot as plt
```

```python
query = """SELECT 
    s.specobjid, s.ra, s.dec, g.h_alpha_flux, g.h_beta_flux,
    g.oiii_5007_flux, g.nii_6548_flux, g.nii_6584_flux,
    g.sii_6717_flux, g.sii_6731_flux, g.oi_6300_flux, 
    ID.bptclass
    
    FROM SpecObjAll AS s
    JOIN galSpecExtra AS ID ON ID.specObjID = s.specObjID
    JOIN galSpecLine AS g ON g.specObjID = s.specObjID
    
    WHERE NOT h_alpha_flux <='0' AND NOT h_beta_flux <='0' 
    AND NOT nii_6548_flux <='0' AND NOT nii_6584_flux <='0'
    AND NOT sii_6717_flux <='0' AND NOT sii_6731_flux <='0'
    AND NOT oi_6300_flux <='0' AND NOT oiii_5007_flux <='0'
    
    ORDER BY s.specobjid ASC"""

data = SDSS.query_sql(query)
data
```

## Razones

```python
ejey_OHb=np.log((data['oiii_5007_flux'])/(data['h_beta_flux']))

ejex_NH=np.log((data['nii_6548_flux']+data['nii_6584_flux'])/data['h_alpha_flux'])
ejex_SH=np.log((data['sii_6717_flux']+data['sii_6731_flux'])/data['h_alpha_flux'])
ejex_OH=np.log(data['oi_6300_flux']/data['h_alpha_flux'])
```

## Gráficas

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(ejex_NH, ejey_OHb, marker=".", color="purple")
# Cuadrado:
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
plt.title('BPT Clásico, Gráfico a')
plt.ylabel('log [O III]/ Hβ')
plt.xlabel('log [NII] /Hα')

# Kewley
x1 = np.linspace(-10,0.4)  
y1 = 1.19 + (0.61/(x1-0.47))
plt.plot(x1, y1, '-', color='red', lw=2, label= 'Kewley') 


# Kauffmann
x2 = np.linspace(-10,0)
y2 = 1.3 + (0.61/(x2-0.05))
plt.plot(x2, y2, '--', color='blue', lw=2, label='Kauffman')


# Seyfert-LINER
x3 = np.linspace(-0.2,5)
y3 = 1.01 * x3 +0.48
plt.plot(x3, y3, '-.', color='green', lw=2, label='Seyfert-LINER')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(ejex_SH,ejey_OHb, marker="+", color="blue")
# Cuadrado:
ax1.set_xlim([-10, 10])
ax1.set_ylim([-10, 10])
plt.title('BPT Clásico, Gráfico b')
plt.ylabel('log [O III]/ Hβ')
plt.xlabel('log [SII] /Hα')

# Kewley
x1 = np.linspace(-10,0.3) 
y1 = 1.30 + (0.72/(x1-0.32))
plt.plot(x1, y1, '-' , color='red', lw=2, label= 'Kewley') 


# Seyfert-LINER
x2 = np.linspace(-0.32,4.0)
y2 = 0.76 + (1.89*x2)
plt.plot(x2, y2, '-.', color='green', lw=2, label='Seyfert-LINER')
```

```python
fig = plt.figure()
ax2 = fig.add_subplot() 
ax2.scatter(ejex_OH, ejey_OHb, marker="*", color="cyan")
# Cuadrado:
ax2.set_xlim([-15, 10])
ax2.set_ylim([-10, 10])
plt.title('BPT Clásico, Gráfico c')
plt.ylabel('log [O III]/ Hβ')
plt.xlabel('log [OI] /Hα')

# Kewley
x1 = np.linspace(-15, -0.6) 
y1 = 1.33 + (0.73/(x1+0.59))
plt.plot(x1, y1, '-', color='red', lw=2, label= 'Kewley') 


# Seyfert-LINER
x2 = np.linspace(-1.15, 8)
y2 = 1.3 + (1.18*x2) 
plt.plot(x2, y2, '-.', color='green', lw=2, label='Seyfert-LINER')
```

```python
fig, ax = plt.subplots(1,3, figsize=(9,6))

ax[0].scatter(ejex_NH,ejey_OHb, marker=".", color="purple")
ax[0].set_xlabel('NII/Hα')
ax[0].set_ylabel('OIII/Hβ')
ax[0].set_xlim(-10, 10)
ax[0].set_ylim(-10, 10)
# Kewley
x0 = np.linspace(-10,0.4) 
y0 = 1.19 + (0.61/(x1-0.47))
ax[0].plot(x0, y0, '--' , color='red', lw=2, label= 'Kewley') 


ax[1].scatter(ejex_SH,ejey_OHb, marker="+", color="blue")
ax[1].set_xlabel('SII/Hα')
ax[1].set_ylabel('OIII/Hβ')
ax[1].set_xlim(-10, 10)
ax[1].set_ylim(-10, 10)
# Kewley
x1 = np.linspace(-10, 0.3) 
y1 = 1.30 + (0.72/(x1-0.32))
ax[1].plot(x1, y1, '--' , color='red', lw=2, label= 'Kewley') 


ax[2].scatter(ejex_OH,ejey_OHb, marker="*", color="cyan")
ax[2].set_xlabel('OI/Hα')
ax[2].set_ylabel('OIII/Hβ')
ax[2].set_xlim(-10,10)
ax[2].set_ylim(-10,10)
# Kewley
x1 = np.linspace(-15, -0.6) 
y1 = 1.33 + (0.73/(x1+0.59))
plt.plot(x1, y1, '--' , color='red', lw=2, label= 'Kewley') 
```
