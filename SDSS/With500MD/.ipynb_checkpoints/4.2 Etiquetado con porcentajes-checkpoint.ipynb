```python
# Valores en los que en los 3 PreEtiquetados da el mismo valor:
Coin100 = dfTotal[dfTotal['PreEtiquetaNII'] == dfTotal['PreEtiquetaSII']]
Coin100 = Coin100[Coin100['PreEtiquetaNII'] == Coin100['PreEtiquetaOI']]
```

## Etiquetado al 66.7% de coincidencia:

```python
# Función para obtener el valor de las 2 columnas iguales o None
def obtener_valor_igual(row):
    valores = [row['PreEtiquetaNII'], row['PreEtiquetaSII'], row['PreEtiquetaOI']]
    if valores.count(valores[0]) == 2:
        return valores[0]
    else:
        return None
```

```python
# Aplicar la función a cada fila y crear la columna "Coin66%"
dfTotal['Coin66%'] = dfTotal.apply(obtener_valor_igual, axis=1)
```

```python
dfTotal[52:57]
```

```python
# Selecciona las columnas deseadas
columnas_elegidas = ['specobjid', 'ra', 'dec', 'z', 'h_alpha_flux', 'h_beta_flux', 'oiii_5007_flux', 'nii_6548_flux', 'nii_6584_flux', 'sii_6717_flux', 'sii_6731_flux', 'oi_6300_flux', 'subclass', 'Ejey_OHB', 'Ejex_NH', 'Ejex_SH', 'Ejex_OH', 'Coin66%']

# Crea un nuevo DataFrame con las columnas deseadas
Coin66 = dfTotal[columnas_elegidas]
Coin66[100:105]
```

```python
# Conteo
conteo_66 = Coin66['Coin66%'].value_counts()
conteo_66
```
