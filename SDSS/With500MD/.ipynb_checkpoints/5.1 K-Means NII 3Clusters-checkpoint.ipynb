## 1. Hacer K-Means para X = 'Ejex_NH', 'Ejey_OHB'.

```python
X = dfTotal[['Ejex_NH', 'Ejey_OHB']]
```

```python
# Llamo al metodo con 3 Clusters
kmeans = KMeans(n_clusters=3)
```

```python
# Entreno el modelo con mis X
kmeans.fit(X)
```

```python
# Agregar la columna "Cluster" a la tabla data2.
dfTotal['ClusterN3'] = kmeans.labels_
dfTotal[0:5]
```

## 2. Realizar grafica de clusters con los colores de la gráfica "3.2 Pre-Etiqueta NII 4GRUPOS".

```python
# Crear un diccionario que mapee los valores de kmeans.labels_ a los colores correspondientes
colors_dict = {
    0: "indigo", 
    1: "salmon", 
    2: "lightsalmon", 
}

# Crear una lista de colores para cada punto en la gráfica
colors = [colors_dict[label] for label in kmeans.labels_]

# Graficar los puntos con los colores correspondientes
plt.scatter(X['Ejex_NH'], X['Ejey_OHB'], c=colors)


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='red', marker='X', label='Centroids') #Centroides de x y centroides de y.
plt.xlabel('Ejex_NH')
plt.ylabel('Ejey_OHB')
plt.title('KMeans Ejex_NH vs Ejey_OHB \n clusters=3')
plt.legend()
plt.show()
```

Se puede comparar con la imagen de 3.2 Pre-Etiqueta NII 4GRUPOS

## 4. Cruzar los dataframes dfTotal Vs dfref.

```python
# Contar las filas donde los valores en ambas columnas son iguales para "Infrarroja"
igual_infrarroja = ((dfTotal["ClusterN3"] == "Infrarroja") & (dfTotal["MuestraEtiq"] == "Infrarroja")).sum()

# Contar las filas donde los valores en ambas columnas son iguales para "Seyfert"
igual_seyfert = ((dfTotal["ClusterN3"] == "Seyfert") & (dfTotal["MuestraEtiq"] == "Seyfert")).sum()

# Contar las filas donde los valores en ambas columnas son iguales para "LINER"
igual_liner = ((dfTotal["ClusterN3"] == "LINER") & (dfTotal["MuestraEtiq"] == "LINER")).sum()
```
